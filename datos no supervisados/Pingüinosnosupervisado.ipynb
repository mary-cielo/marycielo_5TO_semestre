```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
```

```python

np.random.seed(42)
data = np.random.rand(100, 2)
data
```

```python
kmeans= KMeans(n_clusters=5, random_state=42)
kmeans.fit(data)
```

```python
centroide = kmeans.cluster_centers_
centroide
```

```python
etiquetas = kmeans.labels_
etiquetas
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data[:,0], data[:,1], c=etiquetas, cmap='viridis', marker='o')
plt.scatter(centroide[:,0], centroide[:,1], c='red', marker='X', s=200, label='Centroides')
plt.title('visualizacion de K-Means ')
plt.legend()
```

```python
pinguinos = sns.load_dataset('penguins')
pinguinos
```

```python
data = pinguinos[['bill_length_mm', 'bill_depth_mm']].dropna()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data)
```

```python
centroide = kmeans.cluster_centers_
centroide
```

```python
etiquetas = kmeans.labels_
etiquetas
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', hue=etiquetas, palette='viridis', data=data, s=100)
plt.scatter(centroide[:,0], centroide[:,1], c='red', marker='X', s=200, label='Centroides')
plt.title('K-Means en datos de pingüinos')    
plt.legend()  
```
