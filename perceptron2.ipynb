```python
# definicion de datos de entrenamiento 
datos=[
    [500, 1, 1],
    [50, 0, 0],
    [650, 1, 0],
    [10, 1, 1]
]

etiquetas= [0,1,0,1 ]
```

```python
# normalizar datos 
minutos_crudos = [fila[0]for fila in datos]
minutos_min = min(minutos_crudos)
minutos_max = max(minutos_crudos)

def normalizar(valor, min_valor, max_valor):
    return (valor - min_valor) / (max_valor - min_valor)

datos_normalizados=[]
for fila in datos:
    min_norm= normalizar(fila[0], minutos_min, minutos_max)
    asistencia= fila[1]
    participacion = fila [2]
    datos_normalizados.append([min_norm,asistencia, participacion])
    
print(datos_normalizados)
```

```python
# inizializar parametros 
w1=0.0
w2=0.0
w3=0.0
bias=0.0
taza_aprendizaje=0.1
```

```python
#funcion de activacion
def  escalar (valor):
    return 1 if valor>=0 else 0 
```

```python
#entrenamineto 
for epoca in range (20):
    print (f"\n Epoca {epoca+1}")
    for i in range (len(datos_normalizados)):
        x1,x2,x3=datos_normalizados[i]
        salida_deseada = etiquetas[i]
    
        z=x1*w1+x2*w2+bias
        salida_obtenida = escalar(z)
        
        print(f"\n entrada: minutos= {x1:.2f} asistencia={x2} salida:{salida_obtenida}, esperada:{salida_deseada}")
        error= salida_deseada-salida_obtenida
        
        w1+= taza_aprendizaje*error
        w2+= taza_aprendizaje*error
        bias += taza_aprendizaje
        print(f"w1={w1:.4f}, w2={w2:.4f}, bias={bias:.4f}")
```

```python
def predecir (minutos, participacion, asistencia):
    min_norm = normalizar(minutos,minutos_max, minutos_min)
    z= minutos_min*w1+asistencia*w2+ participacion*w3+bias
    return escalar(z)
print (f"\n-- pruebas")
print ("desaprobo" if predecir(10,0,0) else "aprobo")
print ("desaprobo"if predecir(500,0,1)else "aprobo")
```
