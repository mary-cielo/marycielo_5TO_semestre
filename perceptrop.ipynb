```python
# definicion de datos de entrenamiento 
datos=[
    [150, 1],
    [2500, 1],
    [160, 0],
    [2600, 0]
]

etiquetas= [0,1,0,1 ]
```

```python
# normalizar datos 
pesos_crudos = [fila[0]for fila in datos]
pesos_min = min(pesos_crudos)
pesos_max = max(pesos_crudos)

def normalizar(valor, min_valor, max_valor):
    return (valor - min_valor) / (max_valor - min_valor)

datos_normalizados=[]
for fila in datos:
    peso_norm= normalizar(fila[0], pesos_min, pesos_max)
    color= fila[1]
    datos_normalizados.append([peso_norm,color])
    
print(datos_normalizados)
```

```python
# inizializar parametros 
w1=0.0
w2=0.0
bias=0.0
taza_aprendizaje=0.1
```

```python
# funcion de activacion 
def  escalar (valor):
    return 1 if valor>=0 else 0 
```

```python
# entrenamiento 
for epoca in range (20):
    print (f"\n Epoca {epoca+1}")
    for i in range (len(datos_normalizados)):
        x1,x2=datos_normalizados[i]
        salida_deseada = etiquetas[i]
    
        z=x1*w1+x2*w2+bias
        salida_obtenida = escalar(z)
        
        print(f"\n eentrada: peso= {x1:.2f} color={x2} salida:{salida_obtenida}, esperada:{salida_deseada}")
        error= salida_deseada-salida_obtenida
        
        w1+= taza_aprendizaje*error
        w2+= taza_aprendizaje*error
        bias += taza_aprendizaje
        print(f"w1={w1:.4f}, w2={w2:.4f}, bias={bias:.4f}")
```

```python
def predecir(peso,color):
    peso_norm= normalizar(peso, pesos_min,pesos_max)
    z= peso_norm*w1+color*w2+bias
    return escalar(z)
print (f"\n-- prueab con nuevas frutas")
print ("fruta con 140 gramos y color rojo", "sandia" if predecir(140,1) else "manzana")
print ("ruta con 3000 gramos y color verde", "sandia"if predecir(3000,0)else "manzana")
```
