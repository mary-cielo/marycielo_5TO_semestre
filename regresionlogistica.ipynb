```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("datos_seguro.csv")
df.head()
```

```python
plt.scatter(df.edad, df.compra)
```

```python
X_entrena, X_prueba, y_entrena, y_prueba=train_test_split(df[["edad"]],df[["compra"]], train_size=0.8, random_state=36)
```

```python
modelo=LogisticRegression()
```

```python
modelo.fit(X_entrena, y_entrena)
```

```python
modelo.score(X_entrena,X_entrena)
```

```python
datos_nuevos= pd.DataFrame({"edad":[25,35,75,85]})
datos_nuevos
```

```python
probabilidad= modelo.predict_proba(datos_nuevos)

probabilidad
```

```python
probabilidad_compra = probabilidad [:,1]
probabilidad_compra
```

```python
plt.scatter(df.edad, df.compra)
plt.scatter(datos_nuevos["edad"],probabilidad_compra,color="red")
```
